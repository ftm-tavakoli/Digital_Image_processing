```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage.io import imread
import cv2
import sys
import warnings
```

```python
    if not sys.warnoptions:
        warnings.simplefilter("ignore")
```

```python
im = imread('biden_fall_3.png')
```

```python
imgplot = plt.imshow(im,cmap="gray") # Displaying the image
plt.title('Input Image')
plt.show()
```

```python
kernel = np.ones((9,9), np.uint8)
```

```python

img_erosion_1 = cv2.erode(im[:,:,0], kernel, iterations=1)
img_erosion_array_1=np.array(img_erosion_1 ,dtype=np.float)
img_dilation_1 = cv2.dilate(im[:,:,0], kernel, iterations=1)
img_dilation_array_1=np.array(img_dilation_1,dtype=np.float)

img_erosion_2 = cv2.erode(im[:,:,1], kernel, iterations=1)
img_erosion_array_2=np.array(img_erosion_2 ,dtype=np.float)
img_dilation_2 = cv2.dilate(im[:,:,1], kernel, iterations=1)
img_dilation_array_2=np.array(img_dilation_2,dtype=np.float)

img_erosion_3 = cv2.erode(im[:,:,2], kernel, iterations=1)
img_erosion_array_3=np.array(img_erosion_3 ,dtype=np.float)
img_dilation_3 = cv2.dilate(im[:,:,2], kernel, iterations=1)
img_dilation_array_3=np.array(img_dilation_3,dtype=np.float)
```

```python

img_erosion = cv2.merge([img_erosion_1, img_erosion_2, img_erosion_3])
img_dilation = cv2.merge([img_dilation_1, img_dilation_2, img_dilation_3])
```

```python
imgplot=plt.imshow(img_erosion,cmap="gray")
plt.title('Erosion')
plt.show()
imgplot=plt.imshow(img_dilation,cmap="gray")
plt.title('Dilation')
plt.show()
```

```python

of dilation and erosion ====> smoothing
img_smooth_array_1 = (img_erosion_array_1+img_dilation_array_1)/2
img_smooth_1= np.array(img_smooth_array_1 ,dtype= np.uint8)

img_smooth_array_2 = (img_erosion_array_2+img_dilation_array_2)/2
img_smooth_2= np.array(img_smooth_array_2 ,dtype= np.uint8)

img_smooth_array_3 = (img_erosion_array_3+img_dilation_array_3)/2
img_smooth_3= np.array(img_smooth_array_3 ,dtype= np.uint8)

img_smooth = cv2.merge([img_smooth_1, img_smooth_2, img_smooth_3])
imgplot=plt.imshow(img_smooth,cmap="gray")
plt.title('After Smoothing')
plt.show()
```
