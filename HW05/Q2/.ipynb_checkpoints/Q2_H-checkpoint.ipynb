```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage.io import imread
import cv2
import sys
import warnings
```

```python
if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
im = imread('biden_fall_3.png')
```

```python
imgplot = plt.imshow(im,cmap="gray") # Displaying the image
plt.title('Input Image')
plt.show()
```

```python
kernel = np.ones((9,9), np.uint8)
```

```python

img_open_1 = cv2.morphologyEx(im[:,:,0], cv2.MORPH_OPEN, kernel)
img_open_array_1 = np.array(img_open_1 ,dtype=np.float)
img_close_1 = cv2.morphologyEx(im[:,:,0], cv2.MORPH_CLOSE, kernel)
img_close_array_1 = np.array(img_close_1 ,dtype=np.float)

img_open_2 = cv2.morphologyEx(im[:,:,1], cv2.MORPH_OPEN, kernel)
img_open_array_2 = np.array(img_open_2 ,dtype=np.float)
img_close_2 = cv2.morphologyEx(im[:,:,1], cv2.MORPH_CLOSE, kernel)
img_close_array_2 = np.array(img_close_2 ,dtype=np.float)


img_open_3 = cv2.morphologyEx(im[:,:,2], cv2.MORPH_OPEN, kernel)
img_open_array_3 = np.array(img_open_3 ,dtype=np.float)
img_close_3 = cv2.morphologyEx(im[:,:,2], cv2.MORPH_CLOSE, kernel)
img_close_array_3 = np.array(img_close_3 ,dtype=np.float)
```

```python

img_open = cv2.merge([img_open_1, img_open_2, img_open_3])
img_close = cv2.merge([img_close_1, img_close_2, img_close_3])
```

```python
imgplot=plt.imshow(img_open,cmap="gray")
plt.title('Opening')
plt.show()
imgplot=plt.imshow(img_close,cmap="gray")
plt.title('Closing')
plt.show()
```

```python

img_smooth_array_1 = (img_open_array_1+img_close_array_1)/2
img_smooth_1= np.array(img_smooth_array_1 ,dtype= np.uint8)

img_smooth_array_2 = (img_open_array_2+img_close_array_2)/2
img_smooth_2= np.array(img_smooth_array_2 ,dtype= np.uint8)

img_smooth_array_3 = (img_open_array_3+img_close_array_3)/2
img_smooth_3= np.array(img_smooth_array_3 ,dtype= np.uint8)

img_smooth = cv2.merge([img_smooth_1, img_smooth_2, img_smooth_3])
imgplot=plt.imshow(img_smooth,cmap="gray")
plt.title('After Smoothing')
plt.show()
```
