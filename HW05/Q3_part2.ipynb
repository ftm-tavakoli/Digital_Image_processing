```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib.pyplot as plt
import glob
import math
```

# F & J

```python
images = [cv2.imread(file) for file in glob.glob(r"C:\Users\fateme\Desktop\P3\objects\*.png")]
```

```python
objects = ['tuna   ','persil ','soda   ','milk   ','tiny   ','crest  ','cake   ','oil    ','maxi   ','shampoo','tea    ','chips  ']
multiply =[2,1,1,1,2,1,2,1,2,1,1,2]
price = [40000, 83000,6500,30000,5000,50000,8000,108000,6000,139000,144000,36000]
count = 0
pay = 0 

print('name    number      price')
print('---------------------------------')

for i in range(0,len(images)):
    img1 = cv2.imread(r"C:\Users\fateme\Desktop\P3\cart_2.png")
    img2 = images[i]
    match =0
    orb = cv2.ORB_create(nfeatures=1200)
    kp1, des1 = orb.detectAndCompute(img1, None)
    kp2, des2 = orb.detectAndCompute(img2, None)

    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    matches = bf.match(des1, des2)

    matches = sorted(matches, key=lambda x: x.distance)

    match_img = cv2.drawMatches(img1, kp1, img2, kp2, matches[:10], None)
    match += 1
    print(objects[i] +'    ' +str((match)*multiply[i]) +'        '+ str(price[i]))
    count += (match*multiply[i])
    pay += match*multiply[i]*price[i]
#     cv2.waitKey()
print('------------------------------')
print('total number of items : '+ str(count))
print('cost of all items eqal to : '+ str( pay))
```
