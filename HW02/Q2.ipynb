```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
```

```python
def bitplane_slice(image, plot=False):
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    bitplane=[]
    for i in range(8):
        bitplane.append((image_gray // 2 ** i) % 2)
   
    if plot:
        fig, ax = plt.subplots(2, 4,figsize=(12,6))
        for i in range(8):
            ax[i//4, i-4*(i//4)].imshow(bitplane[i],cmap='gray')
            ax[i//4, i-4*(i//4)].set_title('k =%s '%i)
            plt.grid(False)
    
    plt.show()
    
    return np.array(bitplane)
    
```

```python
def xor(image1, image2, plot=False):
    bit_plane_image1 = bitplane_slice(image1)
    bit_plane_image2 = bitplane_slice(image2)

    bit_plane_xor = np.logical_xor(bit_plane_image1,bit_plane_image2)
    
    if plot:
        fig, ax = plt.subplots(2, 4,figsize=(12,8))
        for i in range(8):
                ax[i//4, i-4*(i//4)].imshow(bit_plane_xor[i],cmap='gray')
                ax[i//4, i-4*(i//4)].set_title('k =%s '%i)
            
    plt.show()
    return bit_plane_xor
```

```python
f1_01 = cv2.imread('P2//football_f1_01.png')
f1_02 = cv2.imread('P2//football_f1_02.png')
xor(f1_01 ,f1_02 , True)
```

```python
f3_01 = cv2.imread('P2//football_f3_01.png')
f3_02 = cv2.imread('P2//football_f3_02.png')
xor(f3_01 ,f3_02 , True)
```

```python
def moving_frames(image1, image2, plot = False):
    bit_plane_xor = xor(image1, image2)
    moving = 0
    for i in range(4,8):
        moving += bit_plane_xor[i] * 2 ** i

    if plot:
        plt.figure()
        plt.imshow(moving, cmap='gray')
        plt.show()
        
    return moving
```

```python
f1_01 = cv2.imread('P2//football_f1_01.png')
f1_02 = cv2.imread('P2//football_f1_02.png')
moving_frames(f1_01 ,f1_02 , True)
```

```python
def enhancement(image):

    img = cv2.GaussianBlur(image.astype('float32'), (9, 9), 0)
    _, enhanced = cv2.threshold(img.astype('float32'), 30, 255, cv2.THRESH_BINARY)
    plt.figure()
    plt.imshow(enhanced, cmap="gray")
    plt.show()
```

```python
f1_01 = cv2.imread('P2//football_f1_01.png')
f1_02 = cv2.imread('P2//football_f1_02.png')
mf1 = moving_frames(f1_01 ,f1_02 , False)
enhancement(mf1)
```
